```python
X = df.drop('target', axis = 1)
```

```python
X
```

```python
y = df['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.10, random_state = 101)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test) # we need to transform X_test to avoid data leakage from the X_test.
```

```python
from sklearn.linear_model import LogisticRegressionCV # Using RegressionCV
```

```python
help(LogisticRegressionCV) # Cs=10(NOTE)
```

```python
log_model = LogisticRegressionCV()
```

```python
log_model.fit(scaled_X_train, y_train)
```

```python
log_model.coef_
```

```python
log_model.C_ # C valuw with the best fit based of its cross validation
```

```python
log_model.Cs_ # all the c values that were checked
```

```python
log_model.get_params() # to get the paramters that were provided.
```

```python
# To compare the coefs to the features from the data.

coefs = pd.Series(index = X.columns, data = log_model.coef_[0])
```

```python
coefs
```

```python
sns.barplot(x = coefs.index, y = coefs.values)
plt.show()
```

```python
# To sort:
coefs = coefs.sort_values()
plt.figure(figsize =(10, 6), dpi = 130)
sns.barplot(x = coefs.index, y = coefs.values)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python

plot_confusion_matrix(log_model, scaled_X_test, y_test)
plt.show()
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import plot_precision_recall_curve, plot_roc_curve
```

```python
fig, ax = plt.subplots(figsize =(8, 4), dpi = 120)
plot_precision_recall_curve(log_model, scaled_X_test, y_test, ax = ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize =(8, 4), dpi = 120)
plot_roc_curve(log_model, scaled_X_test, y_test, ax = ax)
plt.show()
```
